```python
import pandas as pd

url = "https://raw.githubusercontent.com/stephenfrein/vucsc8491/refs/heads/main/LifeExpectancyData.csv"
life_df = pd.read_csv(url)

life_df.head()
```

```python
print("Rows:", life_df.shape[0])
print("Columns:", life_df.shape[1])
```

```python
null_pct = (life_df.isnull().sum() / len(life_df)) * 100
null_pct.sort_values(ascending=False)
```

```python
life_df[
    (life_df['Country'].str.contains('United', na=False)) &
    (life_df['Status'] == 'Developing')
]
```

```python
life_df[life_df['Life expectancy '] > 87]
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

corr = life_df.select_dtypes(include=np.number).corr()

plt.figure(figsize=(12,10))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
corr['Life expectancy'].sort_values(ascending=False)
```
